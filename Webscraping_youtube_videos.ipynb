```python
from pytube import YouTube
from pytube import Search
from PIL import Image
from IPython.display import display
import requests
from io import BytesIO
```

```python
se=str(input("What to search: "))
limit=int(input("Limit: "))
s=Search(se)
v=0
for video in s.results[:limit]:
  print('\033[1m' +'Video Number:'+ '\033[0m',v+1)
  v+=1
  print("Title: ",video.title)
  th=video.thumbnail_url
  response = requests.get(th)
  image = Image.open(BytesIO(response.content))
  image = image.resize((128, 128))
  display(image)
  print("url: ",video.embed_url)
  yt=YouTube(video.embed_url)
  print("Title: ",yt.title)
  print("Views: ",yt.views)
  print("Description: ",yt.description)
  print("Author: ",yt.author)
  print("Age restricted: ",yt.age_restricted)
  print("Ratings: ",yt.rating)
  print("Duration: ",yt.length)
  print("Channel id: ",yt.channel_url)
  print("Publish date: ",yt.publish_date)
```
